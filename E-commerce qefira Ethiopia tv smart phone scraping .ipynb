```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
name = []
condition = []
location = []
price = []
for website in range(1,20):
    url = f"https://www.qefira.com/vans-buses?page="
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}     
    response = requests.get(url,headers = headers)

    soup = BeautifulSoup(response.content, 'html.parser')
    soup1 = BeautifulSoup(soup.prettify(), 'html.parser')

    page1 = soup1.find_all('div',{'class':"listing-card__content 1"})
  
    for cars in page1:

        #name
        try:
            name.append(cars.find('div',{'class':"listing-card__header__title"}).get_text(strip = True))
        except:
                 name.append('n/a')
        #price

        try:
            condition.append(cars.find('div',{'class':"listing-card__header__tags"}).get_text( strip = True))
        except:
                  condition.append('n/a')
        try:
            location.append(cars.find('div',{'class':"listing-card__header__location"}).get_text( strip = True))
        except:
                  location.append('n/a')
        try:
            price.append(cars.find('span',{'class':"listing-card__price"}).get_text(strip = True))
        except:
                  price.append('n/a')
```

```python
ca = pd.DataFrame({'name':name,'condition':condition,'location':location,'price':price,})
```
